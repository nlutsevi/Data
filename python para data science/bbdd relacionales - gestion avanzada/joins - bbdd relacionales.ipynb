```python
import pandas as pd
consumidores = [("A", "Movil"), ("B", "Movil"), ("A", "Portatil"), 
               ("A", "Tablet"), ("B", "Tablet"), ("C", "Portatil"),
               ("D", "Smartwatch"), ("E", "Consola")]
con_labels = ["Consumidor", "Producto"]

con_df = pd.DataFrame.from_records(consumidores, columns = con_labels)
```

```python
productores = [("a", "Movil"), ("a","Smartwatch"), ("a", "Tablet"),
              ("b", "Portatil"), ("c", "Sobremesa"), ("c", "Portatil")]

prod_labels = ["Productor", "Producto"]

prod_df = pd.DataFrame.from_records(productores, columns = prod_labels)
```
